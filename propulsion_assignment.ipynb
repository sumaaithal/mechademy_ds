```python
#importing libaries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#reading the csv file
df = pd.read_csv("C:\\Users\\USER\\Desktop\\ds\\propulsion.csv",index_col=0)
```

```python
df.head()
```

```python
df.shape
```

```python
#copying the df to other variable
prop = df.copy()
```

```python
#checking the columns
prop.columns
```

```python
#model1 using multiple linear regression 
```

```python
#pred of GT Compressor decay state coefficient
```

```python
x_features = prop.drop(['GT Compressor decay state coefficient.','GT Turbine decay state coefficient.'],axis=1)
y = df['GT Compressor decay state coefficient.']
```

```python
X = x_features
```

```python
#splitting the data set to train and test
x_train,x_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=0)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
#pred of GT Compressor decay state coefficient.
lm_model = LinearRegression()
lm_model.fit(x_train,y_train)
y_pred = lm_model.predict(x_test)
print(np.sqrt(mean_squared_error(y_pred,y_test)))
print(r2_score(y_pred,y_test)*100)
```

```python
#pred of GT Turbine decay state coefficient
```

```python
X2 = x_features
y1 = df['GT Turbine decay state coefficient.']
```

```python
x1_train,x1_test,y1_train,y1_test = train_test_split(X2,y1,train_size=0.8,random_state=0)
```

```python
lm_model1 = LinearRegression()
lm_model1.fit(x1_train,y1_train)
y1_pred = lm_model1.predict(x1_test)
print(np.sqrt(mean_squared_error(y1_pred,y1_test)))
print(r2_score(y1_pred,y1_test)*100)
```

```python
#model2 using decsion tree regressor for GT Compressor decay state coefficient.
```

```python
X3 = x_features
y = df['GT Compressor decay state coefficient.']
```

```python
x2_train,x2_test,y2_train,y2_test = train_test_split(X3,y,train_size=0.8,random_state=0)
```

```python
dt_model = DecisionTreeRegressor()
dt_model.fit(x2_train,y2_train)
y_pred_dt = dt_model.predict(x2_test)
print(np.sqrt(mean_squared_error(y_pred_dt,y2_test)))
print(r2_score(y_pred_dt,y2_test)*100)
```

```python
x3_train,x3_test,y3_train,y3_test = train_test_split(X4,y,train_size=0.8,random_state=0)
```

```python
#hyperparameter tuned decision tree regressor for GT Compressor decay state coefficient
```

```python
tuned_params = [{'criterion':['mse','mae'],'max_depth':range(2,10),'max_features':['auto','sqrt','log2']}]
```

```python
reg_dt = DecisionTreeRegressor()
```

```python
dt_model2 = GridSearchCV(reg_dt,param_grid=tuned_params,cv=10,scoring='r2',verbose=3)
```

```python
X5 = x_features
y = df['GT Compressor decay state coefficient.']
```

```python
x4_train,x4_test,y4_train,y4_test = train_test_split(X5,y,train_size=0.8,random_state=0)
```

```python
dt_model2.fit(x4_train,y4_train)
```

```python
print(dt_model2.best_score_)
print(dt_model2.best_params_)
```

```python
dt_model3 = DecisionTreeRegressor(criterion='mse',max_depth=9,max_features='auto')
```

```python
dt_model3.fit(x4_train,y4_train)
dt_model3_pred = dt_model3.predict(x4_test)
print(np.sqrt(mean_squared_error(dt_model3_pred,y4_test)))
print(r2_score(dt_model3_pred,y4_test)*100)
```
